Outlier Detection

```python
filtered_df = df1[(df1['Thermal sensation'] >= -3) & (df1['Thermal preference'] == 'cooler')& (df1['Thermal comfort'] >= 5)]

count_of_rows_to_remove = len(filtered_df)
df1 = df1.drop(filtered_df.index)
print(f"\nCount of rows removed: {count_of_rows_to_remove}")

```

```python
filtered_df = df1[(df1['Thermal sensation'] >= 3) & (df1['Thermal preference'] == 'warmer')& (df1['Thermal comfort'] >= 5)]
count_of_rows_to_remove = len(filtered_df)
df1 = df1.drop(filtered_df.index)
print(f"\nCount of rows removed: {count_of_rows_to_remove}")
```

```python
filtered_df = df1[((df1['Thermal comfort'] > 4) & (df1['Thermal sensation acceptability'] == '0')) | ((df1['Thermal comfort'] <3) & (df1['Thermal sensation acceptability'] == '1'))]
count_of_rows_to_remove = len(filtered_df)
df1 = df1.drop(filtered_df.index)
print(f"\nCount of rows removed: {count_of_rows_to_remove}")
```

```python
filtered_df = df1[((df1['Thermal sensation'] > 2) & (df1['PMV'] <2)) | ((df1['Thermal sensation'] < -2) & (df1['PMV'] >-2))]
count_of_rows_to_remove = len(filtered_df)
df1 = df1.drop(filtered_df.index)
print(f"\nCount of rows removed: {count_of_rows_to_remove}")
```
